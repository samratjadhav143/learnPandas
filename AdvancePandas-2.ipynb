```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

# Merge the DataFrames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')

print(merged_df)
```

```python

# Extract subsets of the DataFrame
df1 = df[['Name', 'Survived', 'Pclass']][0:5]
df2 = df[['Name', 'Survived', 'Pclass']][5:10]

# Concatenate vertically (stacking rows)
vertical_concatenated = pd.concat([df1, df2]) 
#ignore_index=True

# Concatenate horizontally (side by side)
horizontal_concatenated = pd.concat([df1, df2], axis=1)

print("Vertically Concatenated DataFrame:")
print(vertical_concatenated)

print("\nHorizontally Concatenated DataFrame:")
print(horizontal_concatenated)
```

```python
import pandas as pd

# Create the 'employees' DataFrame
employees_data = {'EmployeeID': [1, 2, 3, 4, 5],
                  'Name': ['Samrat', 'Bob', 'Charlie', 'David', 'Emma'],
                  'Department': ['HR', 'IT', 'Finance', 'IT', 'HR']}
employees = pd.DataFrame(employees_data)

# Create the 'salaries' DataFrame
salaries_data = {'EmployeeID': [2, 3, 5],
                 'Salary': [60000, 70000, 55000]}
salaries = pd.DataFrame(salaries_data)

# Merge the DataFrames based on 'EmployeeID'
merged_data = pd.merge(employees, salaries, on='EmployeeID', how='left')

print(merged_data)
```
